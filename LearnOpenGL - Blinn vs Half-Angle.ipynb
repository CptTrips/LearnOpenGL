```python
from matplotlib import cm
from matplotlib import pyplot as plt
import numpy as np
```

```python
angle_range = np.linspace(0, np.pi/2)
theta_grid, phi_grid = np.meshgrid(angle_range, angle_range)
```

```python
def blinn_specular(theta, phi):
    return 2*np.sin(theta)*np.sin(phi) - np.cos(theta + phi)

def half_angle_specular(theta, phi):
    mod_o_plus_i = np.sqrt(2 - 2*np.cos(theta, phi)) + 1e-6
    return (np.sin(theta) + np.sin(phi)) / mod_o_plus_i
```

```python
blinn_values = blinn_specular(theta_grid, phi_grid)
half_angle_values = half_angle_specular(theta_grid, phi_grid)
```
